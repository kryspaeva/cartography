```python
import geopandas as gpd
```

```python
kg = gpd.read_file("updated_kg.json")
```

```python
kg['Affected in %'] = ((kg['Respiratory diseases'] / kg['Population']) * 100).round(2)
kg
```

```python
kg.explore(column="Region",  # make choropleth based on "BoroName" column
    tooltip=["Region", "Population", "Respiratory diseases", "Affected in %"],  # show "BoroName" value in tooltip (on hover)
    popup=True,  # show all values in popup (on click)
    tiles="CartoDB positron",  # use "CartoDB positron" tiles
    cmap="Set1",  # use "Set1" matplotlib colormap
    style_kwds=dict(color="black"),)
```
